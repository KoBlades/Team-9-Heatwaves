```python
# Convert the list of dictionaries to a DataFrame
df_economic_data = pd.DataFrame(economic_data)

# Display the DataFrame to confirm it has been loaded correctly
df_economic_data.head()
```

```python
inflation_dropped_columns = df_economic_data[['Year', 'Inflation Rate']]
inflation_dropped_columns.head()
```

```python
# Filter out the rows where Year is 1983 or 1984
inflation_dropped_years = inflation_dropped_columns.query('Year >= 1985')

# Display the updated DataFrame
inflation_dropped_years.head()
```

```python
# Ensure the DataFrame is sorted by 'Year' in ascending order (oldest to newest)
df_economic_data = df_economic_data.sort_values(by='Year')

# Initialize the cumulative inflation starting with 1 for the base year (1985)
df_economic_data['Cumulative_Inflation'] = 1.0

# Iterate over the DataFrame, starting from the second row (index 1) and calculating cumulative inflation
for i in range(1, len(df_economic_data)):
    # Get the inflation rate for the previous year (as a decimal)
    inflation_rate = df_economic_data.iloc[i - 1]['Inflation Rate'] / 100
    
    # Calculate the cumulative inflation using the previous year's cumulative value
    df_economic_data.at[i, 'Cumulative_Inflation'] = df_economic_data.at[i - 1, 'Cumulative_Inflation'] * (1 + inflation_rate)

# Display the DataFrame to check the results
df_economic_data[['Year', 'Inflation Rate', 'Cumulative_Inflation']].head()
```

```python
# Perform an inner merge on the 'Year' column
df_merged = pd.merge(wildfire_costs_dropped_columns, df_economic_data, on='Year', how='inner')

# Display the first few rows of the merged DataFrame
df_merged.head()
```

```python
df_merged = df_merged[['Year', 'Fires', 'Acres', 'Total', 'Inflation Rate', 'Cumulative_Inflation']]
df_merged.head()
```

```python
# Create a new column for inflation-adjusted Total cost
df_merged['Adjusted_Total'] = df_merged['Total'] / df_merged['Cumulative_Inflation']

# Display the updated DataFrame with the new Adjusted_Total column
df_merged[['Year', 'Fires', 'Acres', 'Total', 'Cumulative_Inflation', 'Adjusted_Total']].head()
```
