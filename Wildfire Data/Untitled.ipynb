```python
import pandas as pd

# Skip the first 8 rows and load the relevant data from line 9 onwards
df = pd.read_csv("heatwave_data_nevada.csv", skiprows=8)

# Display the first few rows of the relevant data
df
```

```python
# Filter rows where the "Year" is between 1950 and 2022
df_filtered = df[(df['year'] >= 1950) & (df['year'] <= 2022)]

# Display the filtered data
df_filtered
```

```python
# Check for missing values
print(df_filtered.isnull().sum())
```

```python
# Drop rows with any missing values
df_cleaned = df_filtered.dropna()
```

```python
# Fill NaN values with 0
df_cleaned = df_filtered.fillna(0)
```

```python
print(df_cleaned.isnull().sum())
```

```python
df_cleaned
```

```python
import matplotlib.pyplot as plt

# Plot the line graph
plt.figure(figsize=(10, 6))  # Set the figure size
plt.plot(df_cleaned['year'], df_cleaned['CanESM2 (Average)'], marker='o')

# Add labels and title
plt.xlabel('Year')
plt.ylabel('CanESM2 (Average)')
plt.title('CanESM2 (Average) Over Time (1950-2022)')

# Display the graph
plt.grid(True)
plt.show()
```
